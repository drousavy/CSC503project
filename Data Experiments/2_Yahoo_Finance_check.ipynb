```python
import pandas as pd
from matplotlib import pyplot as plt
```

## yahoo finance DJIA data for 1st dataset

```python
yahoo_djia = pd.read_csv("../Datasets/yahoo-finance/DJI.csv",parse_dates=[0],index_col=0).sort_index()
pd.concat((yahoo_djia.head(3),yahoo_djia.tail(2)))
```

```python
djia = pd.read_csv("../Datasets/djia/upload_DJIA_table.csv",parse_dates=[0],index_col=0).sort_index()
pd.concat((djia.head(3),djia.tail(2)))
```

```python
overlapping_yahoo_djia = yahoo_djia.loc['2008-08-08':'2016-07-01']
pd.concat((overlapping_yahoo_djia.head(3),overlapping_yahoo_djia.tail(2)))
```

```python
date_range = ['2008-08-08','2008-09-30']
series = djia.Close.loc[date_range[0]:date_range[1]]
# Create figure and plot space
fig, ax = plt.subplots(figsize=(12, 3))

# Add x-axis and y-axis
ax.bar(series.index.values,
       series,
       color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Precipitation (inches)",
       title="Daily Total Precipitation\nJune - Aug 2005 for Boulder Creek")

plt.show()
```
